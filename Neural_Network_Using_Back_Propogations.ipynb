```python
#Import numpy and pandas
import numpy as np
import pandas as pd

# Define the Traing data, Test data and NN weghts as global
global XTrain,YTrain,XTest,YTest,Twt
```

```python
#Class is defined to store the nodes values/weights of each layer
class Layer():
    def __init__(self,):
        self.nodes=''
```

```python
#Define an activation function of the NN. This could be Relu or sigmoid function
def Activation(d):
    d=1/(1+np.exp(-d))
    return(d)
```

```python
#Define forward propogation of the NN
def Forward_propogation(D,wt1):
    X[0].nodes=D
    for i in range(len(wt1)):
        D=np.matmul(D,wt1[i].nodes)
        D=Activation(D)
        X[i+1].nodes=D
    return(D[0])
```

```python
#Define the error/cost functon of the the NN
def Cost_function(wt1):
    yp=[]
    for i in range(len(XTrain)):
        yp.append(Forward_propogation(np.array([XTrain[i]]),wt1))
    cost=np.sum((yp-YTrain)**2)
    return(cost)
```

```python
#Train the Model using using
def Back_propogation():
    cost2=0
    yp=[]
    lr=0.5
    while True:
        cost1=Cost_function(Twt)
        if abs(cost1-cost2)/cost1<0.00001:
            break
        else:
            cost2=cost1+0
            for k in range(len(XTrain)):
                a=np.array([Forward_propogation(np.array([XTrain[k]]),Twt)])
                for i in range(len(LN)-1):
                    if i==0:
                        H[len(LN)-2-i].nodes=np.array([YTrain[k]])-a
                        dwt[len(LN)-2-i].nodes=np.matmul(X[len(LN)-2-i].nodes.T,(H[len(LN)-2-i].nodes)*(X[len(LN)-2-i+1].nodes)*(1-X[len(LN)-2-i+1].nodes))
                    else:
                        H[len(LN)-2-i].nodes=np.matmul((H[len(LN)-2-i+1].nodes)*(X[len(LN)-2-i+2].nodes)*(1-X[len(LN)-2-i+2].nodes),(Twt[len(LN)-2-i+1].nodes).T)
                        dwt[len(LN)-2-i].nodes=np.matmul(X[len(LN)-2-i].nodes.T,(H[len(LN)-2-i].nodes)*(X[len(LN)-2-i+1].nodes)*(1-X[len(LN)-2-i+1].nodes))
                for i in range(len(LN)-1):
                    Twt[i].nodes=Twt[i].nodes+lr*dwt[i].nodes
        print(cost1)
```

```python
#Define the NN paramaters
global dwt, Twt, H, X

#Create a dummy data set
XTrain=np.random.random((1000,5))
YTrain=np.sin(np.sum(XTrain,axis=1))

#Define NN layes and nodes in layers
LN=[5,10,10,1]

#Initiate weights(Twt) and derivative(dwt) of the weights
Twt=[]
dwt=[]
H=[]
X=[]
X.append(Layer())
for i in range(len(LN)-1):
    Twt.append(Layer())
    dwt.append(Layer())
    H.append(Layer())
    X.append(Layer())
    Twt[i].nodes=np.random.random((LN[i],LN[i+1]))
    dwt[i].nodes=np.zeros((LN[i],LN[i+1]))
    H[i].nodes=np.zeros((1,LN[i+1]))
```

```python
#Train your Model using back_propogation
Back_propogation()
```
